```python
# ── Names by community ─────────────────────────────────────────────────────
sinhala_first = [
    'Kasun','Nuwan','Dilshan','Chathura','Isuru','Sachith','Lahiru','Tharaka',
    'Malith','Danushka','Ravindu','Chamara','Asela','Dhanushka','Sithum',
    'Kavinda','Buddhika','Thilina','Chanaka','Supun','Gayan','Eranga','Chamath',
    'Hashan','Roshan','Pasan','Udara','Binura','Pathum','Dasun',
    'Sanduni','Dilini','Chathurika','Nimasha','Thilini','Sachini','Harshani',
    'Nadeesha','Chandima','Imesha','Sumudu','Thisari','Sewwandi','Anushka',
    'Malsha','Devindi','Hiruni','Dulani','Chamodi','Ishara'
]
tamil_first = [
    'Arjun','Priya','Kavya','Rahul','Anjali','Vijay','Meena','Kumar',
    'Nithya','Surya','Deepa','Rajan','Lakshmi','Arun','Divya',
    'Siva','Geetha','Karthik','Yamuna','Shankar','Tharini','Senthil',
    'Vimala','Bala','Gayathri','Murugan','Saranya','Dinesh','Kokila','Jegan'
]
muslim_first = [
    'Mohamed','Fathima','Aisha','Hassan','Zainab','Ibrahim','Mariam',
    'Ahmed','Nusrath','Rizwan','Shifana','Imran','Raheema','Ahamed','Rukshana'
]
last_names = [
    'Perera','Silva','Fernando','Jayawardena','Wickramasinghe','Rajapaksa',
    'Gunasekara','Dissanayake','Bandara','Wijesinghe','Seneviratne','Ranasinghe',
    'Karunaratne','Herath','Pathirana','Mendis','Samarasinghe','Weerasinghe',
    'Gunawardena','Jayasuriya','Rathnayake','Kumarasinghe','Fonseka','De Silva',
    'Munasinghe','Senanayake','Liyanage','Madushanka','Rajapakshe','Amarasinghe',
    'Navaratnam','Balasingham','Krishnaswamy','Ratnasabapathy','Shanmuganathan',
    'Selvam','Rasiah','Kandiah','Ponniah','Subramaniam',
    'Farook','Cassim','Ismail','Ahamed','Nazar','Razik','Sheriff'
]

def sri_lankan_name():
    group = random.choices(['Sinhala','Tamil','Muslim'], weights=[70,20,10])[0]
    first = random.choice(sinhala_first if group=='Sinhala' else tamil_first if group=='Tamil' else muslim_first)
    return f'{first} {random.choice(last_names)}'

# ── Universities ───────────────────────────────────────────────────────────
universities = [
    'University of Colombo','University of Moratuwa','University of Peradeniya',
    'University of Kelaniya','University of Sri Jayewardenepura','University of Ruhuna',
    'University of Jaffna','Sabaragamuwa University','Eastern University',
    'South Eastern University','Wayamba University','Uva Wellassa University',
    'SLIIT','NSBM Green University','Informatics Institute of Technology (IIT)',
    'APIIT Sri Lanka','ICBT Campus','British College of Applied Studies',
    'Sri Lanka Institute of Information Technology','Open University of Sri Lanka'
]

# ── Companies by domain ────────────────────────────────────────────────────
companies = {
    'Technology': [
        'WSO2','Sysco LABS','hSenid Business Solutions','IFS R&D International',
        'Virtusa','Codegen International','Pearson Lanka','Zone24x7',
        'Calcey Technologies','Axiata Digital Labs','Dialog Axiata','SLT Mobitel',
        'Rootcode Labs','99X Technology','Arimac Lanka','Enactor','Surge Global'
    ],
    'Finance': [
        'Commercial Bank of Ceylon','Bank of Ceylon','Sampath Bank',
        'Hatton National Bank (HNB)','Nations Trust Bank','Seylan Bank',
        "People's Bank",'NDB Bank','Pan Asia Bank','DFCC Bank',
        'Cargills Bank','Softlogic Finance','LB Finance','LOLC Finance'
    ],
    'Marketing': [
        'Leo Burnett Sri Lanka','Wunderman Thompson Sri Lanka','JWT Sri Lanka',
        'MullenLowe Sri Lanka','Unilever Sri Lanka','Nestlé Lanka',
        'Ceylon Tobacco Company','John Keells Holdings','Hemas Holdings',
        'Aitken Spence','Brandix Lanka','Cargills (Ceylon)','Dilmah Tea'
    ],
    'HR': [
        'John Keells Holdings','MAS Holdings','Brandix Lanka','Hemas Holdings',
        'Aitken Spence','Softlogic Holdings','Cargills (Ceylon)','Expolanka Holdings',
        'Dialog Axiata','Commercial Bank of Ceylon','Sampath Bank','Virtusa'
    ],
    'Operations': [
        'Expolanka Holdings','Aitken Spence Shipping','Lanka Logistics',
        'John Keells Logistics','CMA CGM Lanka','Maersk Lanka',
        'Sri Lanka Ports Authority','Airport & Aviation Services',
        'SriLankan Airlines','Hayleys Advantis','Tokyo Cement','Holcim Lanka'
    ]
}

# ── Locations ──────────────────────────────────────────────────────────────
city_locations = [
    'Colombo','Kandy','Galle','Negombo','Kurunegala','Jaffna','Matara',
    'Anuradhapura','Ratnapura','Badulla','Batticaloa','Trincomalee',
    'Kalutara','Hambantota','Nuwara Eliya','Vavuniya','Ampara','Polonnaruwa'
]
all_locations = city_locations + ['Remote', 'Hybrid']

# ── Contact helpers ────────────────────────────────────────────────────────
email_domains = ['gmail.com','yahoo.com','hotmail.com','outlook.com',
                 'slt.lk','dialog.lk','mobitel.lk','hutch.lk']

def sl_email(name):
    parts = name.lower().split()
    return f'{parts[0]}.{parts[-1]}{random.randint(1,99)}@{random.choice(email_domains)}'

def sl_phone():
    prefixes = ['071','072','074','075','076','077','078','011','031','033',
                '038','041','045','051','054','055','057','063','065','066','067']
    return f'{random.choice(prefixes)}-{random.randint(1000000,9999999)}'

print('Sri Lankan reference data defined')
print(f'   Universities : {len(universities)}')
print(f'   Locations    : {len(all_locations)}')
print(f'   Company pool : {sum(len(v) for v in companies.values())} across 5 domains')
```

## 3. Define Skill Pools by Domain

```python
tech_skills = [
    'Python','R','SQL','Java','JavaScript','C++','C#','Scala','Go','Kotlin',
    'Machine Learning','Deep Learning','NLP','Computer Vision','TensorFlow','PyTorch','Keras','scikit-learn',
    'Data Analysis','Data Visualization','Tableau','Power BI','Excel','Statistics','A/B Testing',
    'Pandas','NumPy','Matplotlib','Seaborn','Plotly',
    'AWS','Azure','GCP','Docker','Kubernetes','CI/CD','Git','Linux',
    'React','Node.js','Django','Flask','FastAPI','REST API','GraphQL',
    'PostgreSQL','MySQL','MongoDB','Redis','Spark','Hadoop','Kafka',
    'Agile','Scrum','Jira','Communication','Problem Solving','Teamwork','Leadership'
]
finance_skills = [
    'Financial Modeling','Excel','VBA','Bloomberg Terminal','Risk Analysis','Portfolio Management',
    'Accounting','Auditing','Tax Planning','IFRS','GAAP','Financial Reporting',
    'Investment Analysis','Equity Research','Derivatives','Fixed Income',
    'SQL','Python','R','Data Analysis','Statistics','Power BI','Tableau',
    'Communication','Presentation Skills','Attention to Detail','Analytical Thinking','Teamwork'
]
marketing_skills = [
    'Digital Marketing','SEO','SEM','Google Analytics','Social Media Marketing',
    'Content Marketing','Email Marketing','HubSpot','Salesforce','CRM',
    'Market Research','Brand Management','Product Marketing','Campaign Management',
    'Adobe Creative Suite','Canva','Copywriting','Communication','Creativity',
    'Data Analysis','Excel','SQL','A/B Testing','Customer Segmentation',
    'Leadership','Teamwork','Project Management'
]
hr_skills = [
    'Recruitment','Talent Acquisition','HR Management','Performance Management',
    'Employee Relations','Training & Development','Compensation & Benefits',
    'HRIS','Workday','SAP HR','Labor Law','Conflict Resolution',
    'Communication','Interpersonal Skills','Organizational Skills','Leadership',
    'Data Analysis','Excel','Reporting','Onboarding','Succession Planning'
]
operations_skills = [
    'Supply Chain Management','Logistics','Procurement','Inventory Management',
    'Process Improvement','Lean','Six Sigma','Project Management','PMP',
    'ERP','SAP','Oracle','Excel','SQL','Data Analysis','Reporting',
    'Quality Assurance','ISO','Risk Management','Vendor Management',
    'Communication','Leadership','Problem Solving','Teamwork'
]

domain_skill_map = {
    'Technology': tech_skills, 'Finance': finance_skills,
    'Marketing':  marketing_skills, 'HR': hr_skills, 'Operations': operations_skills
}

education_levels  = ['High School', 'Diploma', "Bachelor's", "Master's", 'PhD']
experience_levels = ['Entry', 'Junior', 'Mid', 'Senior', 'Lead', 'Manager']
industries        = list(domain_skill_map.keys())
job_types         = ['Full-Time', 'Part-Time', 'Contract', 'Internship', 'Freelance']
exp_years_map     = {'Entry':0,'Junior':1,'Mid':3,'Senior':6,'Lead':9,'Manager':10}

# LKR monthly salary ranges
salary_lkr = {
    'Entry':   (35000,   60000),
    'Junior':  (55000,  100000),
    'Mid':     (90000,  175000),
    'Senior':  (160000, 280000),
    'Lead':    (250000, 400000),
    'Manager': (300000, 550000)
}

print('Skill pools and salary ranges defined (LKR)')
```

## 4. Generate Candidate Profiles (500 Records)

```python
candidates = []
for i in range(1, 501):
    domain     = random.choice(industries)
    pool       = domain_skill_map[domain]
    skills     = random.sample(pool, min(random.randint(4,12), len(pool)))
    exp_level  = random.choice(experience_levels)
    years      = exp_years_map[exp_level] + random.randint(0, 2)
    pref_locs  = random.sample(all_locations, random.randint(1, 3))
    pref_types = random.sample(job_types, random.randint(1, 2))
    name       = sri_lankan_name()
    uni        = random.choice(universities)

    candidates.append({
        'candidate_id':        f'C{i:04d}',
        'name':                name,
        'email':               sl_email(name),
        'phone':               sl_phone(),
        'location':            random.choice(all_locations),
        'education_level':     random.choice(education_levels),
        'university':          uni,
        'field_of_study':      domain + ' related field',
        'years_of_experience': years,
        'experience_level':    exp_level,
        'primary_domain':      domain,
        'skills':              ', '.join(skills),
        'preferred_locations': ', '.join(pref_locs),
        'preferred_job_types': ', '.join(pref_types),
        'summary': (
            f'{exp_level}-level professional with {years} years of experience in '
            f'{domain}. Skilled in {", ".join(skills[:3])}. '
            f'Graduated from {uni}. '
            f'Looking for {pref_types[0]} opportunities in {pref_locs[0]}.'
        )
    })

df_candidates = pd.DataFrame(candidates)
print(f'Generated {len(df_candidates)} candidate profiles')
df_candidates.head(3)
```

## 5. Generate Job Postings (200 Records)

```python
job_titles = {
    'Technology': [
        'Data Scientist','Machine Learning Engineer','Data Analyst','Software Engineer',
        'Backend Developer','Frontend Developer','Full Stack Developer','DevOps Engineer',
        'Cloud Engineer','NLP Engineer','AI Researcher','Business Intelligence Analyst',
        'Data Engineer','Cybersecurity Analyst','Mobile Developer'
    ],
    'Finance': [
        'Financial Analyst','Investment Analyst','Risk Analyst','Accountant',
        'Auditor','Tax Consultant','Portfolio Manager','Credit Analyst',
        'Treasury Analyst','Financial Controller','Compliance Officer'
    ],
    'Marketing': [
        'Digital Marketing Specialist','SEO Analyst','Content Writer','Brand Manager',
        'Marketing Analyst','Social Media Manager','Product Marketing Manager',
        'Campaign Manager','Growth Hacker','Market Research Analyst'
    ],
    'HR': [
        'HR Generalist','Talent Acquisition Specialist','HR Business Partner',
        'Recruiter','Learning & Development Specialist','Compensation Analyst',
        'HR Manager','People Operations Specialist'
    ],
    'Operations': [
        'Operations Analyst','Supply Chain Analyst','Logistics Coordinator',
        'Process Improvement Specialist','Project Manager','Procurement Officer',
        'Quality Assurance Analyst','Operations Manager','ERP Consultant'
    ]
}

today          = datetime.date.today()
six_months_ago = today - datetime.timedelta(days=180)

def random_date(start, end):
    return start + datetime.timedelta(days=random.randint(0, (end - start).days))

jobs = []
for i in range(1, 201):
    domain      = random.choice(industries)
    pool        = domain_skill_map[domain]
    title       = random.choice(job_titles[domain])
    exp_level   = random.choice(experience_levels)
    req_skills  = random.sample(pool, min(random.randint(3,10), len(pool)))
    pref_pool   = [s for s in pool if s not in req_skills]
    pref_skills = random.sample(pref_pool, min(random.randint(1,4), len(pref_pool)))
    sal_min, sal_max = salary_lkr[exp_level]
    sal_lo = random.randint(sal_min, (sal_min+sal_max)//2)
    sal_hi = random.randint((sal_min+sal_max)//2, sal_max)
    company = random.choice(companies[domain])

    jobs.append({
        'job_id':               f'J{i:04d}',
        'job_title':            title,
        'company_name':         company,
        'industry':             domain,
        'location':             random.choice(all_locations),
        'job_type':             random.choice(job_types),
        'experience_level':     exp_level,
        'min_experience_years': exp_years_map[exp_level],
        'education_required':   random.choice(education_levels),
        'required_skills':      ', '.join(req_skills),
        'preferred_skills':     ', '.join(pref_skills),
        'salary_min_lkr':       sal_lo,
        'salary_max_lkr':       sal_hi,
        'posted_date':          str(random_date(six_months_ago, today)),
        'description': (
            f'We are looking for a {exp_level}-level {title} to join our {domain} team at {company}. '
            f'The ideal candidate will have strong skills in {", ".join(req_skills[:3])}. '
            f'You will work on exciting projects and collaborate with cross-functional teams. '
            f'Preferred skills include {", ".join(pref_skills[:2]) if pref_skills else "none"}.'
        )
    })

df_jobs = pd.DataFrame(jobs)
print(f'Generated {len(df_jobs)} job postings')
df_jobs.head(3)
```
